```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By
```

```python
options = Options()
options.add_experimental_option('detach', True)

driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=options)
```

```python
driver.get("https://www.amazon.co.jp")
driver.maximize_window()
```

```python
# Search input
search_input = driver.find_element('css selector', 'input[placeholder="Search Amazon.co.jp"]')
search_input.click()
search_input.send_keys('headphones')
```

```python
# Search start
search_btn = driver.find_element('css selector', 'input[id="nav-search-submit-button"][type="submit"]')
search_btn.click()
```

```python
# Choose by Customer Reviews
%%script false
# star_num = "3"
# css_selector = f'i[class="a-icon a-icon-star-medium a-star-medium-{star_num}"]'
# review_stars = driver.find_element('css selector', css_selector)
star_num = "4 Stars & Up"
css_selector = f'//section[@aria-label="{star_num}"]'
review_stars = driver.find_element(by=By.XPATH, value=css_selector)
review_stars.click()
```

```python
# Back to no filter
%%script false
department_filters = driver.find_elements('css selector', 'span[class="s-back-arrow aok-inline-block"]+span[class="a-size-base a-color-base"]')
for department_filter in department_filters:
    if department_filter.get_attribute("innerHTML") == "Any Department":
        department_filter.click()
        break
```

```python
value = '//div[contains(@data-cel-widget, "search_result_")]'
products = driver.find_elements(by=By.XPATH, value=value)
for product in products:
    print(product.text)
```
