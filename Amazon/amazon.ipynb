```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By
```

```python
options = Options()
options.add_experimental_option('detach', True)

driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=options)
```

```python
driver.get("https://www.amazon.co.jp")
driver.maximize_window()
```

```python
# Search input
search_input = driver.find_element('css selector', 'input[placeholder="Search Amazon.co.jp"]')
search_input.click()
search_input.send_keys('headphones')
```

```python
# Search start
search_btn = driver.find_element('css selector', 'input[id="nav-search-submit-button"][type="submit"]')
search_btn.click()
```

```python
# Choose by Customer Reviews
%%script false
# star_num = "3"
# css_selector = f'i[class="a-icon a-icon-star-medium a-star-medium-{star_num}"]'
# review_stars = driver.find_element('css selector', css_selector)
star_num = "4 Stars & Up"
review_stars_css_selector = f'//section[@aria-label="{star_num}"]'
review_stars = driver.find_element(by=By.XPATH, value=review_stars_css_selector)
review_stars.click()
```

```python
# Back to no filter
%%script false
department_filters = driver.find_elements('css selector', 'span[class="s-back-arrow aok-inline-block"]+span[class="a-size-base a-color-base"]')
for department_filter in department_filters:
    if department_filter.get_attribute("innerHTML") == "Any Department":
        department_filter.click()
        break
```

```python
current_page_btn = driver.find_element(by=By.XPATH, value="//*[contains(@aria-label,  'Current page')]")
current_page_num = int(current_page_btn.text)
print(current_page_num)

end_page_btn = driver.find_element(by=By.XPATH, value='//*[@class="s-pagination-item s-pagination-disabled"]')
end_page_num = int(end_page_btn.text)
print(end_page_num)
```

```python
# Use as func argument
desired_page_num = 5

if desired_page_num > end_page_num:
    exit('Choosing page must be less than max_page')
the_difference = abs(desired_page_num - current_page_num)
# driver.refresh()

products = []
driver.implicitly_wait(3)
if current_page_num < desired_page_num:
    for _ in range(the_difference):
        # Navigated to new page, so need to find element again
        page_products = driver.find_elements(by=By.XPATH, value='//div[@data-component-type="s-search-result"]')
        # print(page_products)
        # There is a lot of trash inside page_products
        for item in page_products:
            result = item.text
            if isinstance(result, str) is False:
                continue
            # print(result)
            products.append(result)
            
        next_page_btn = driver.find_element(by=By.XPATH, value="//*[contains(@aria-label,  'Go to next page')]")
        next_page_btn.click()

driver.implicitly_wait(0)
```

```python
# Print result out
print(len(products))
for product in products:
    print(product)
```
